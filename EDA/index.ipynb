MULTIVARIATE EDA

```python
tips = sns.load_dataset('tips')
titanic= pd.read_csv('train.csv')
flights = sns.load_dataset('flights')
iris = sns.load_dataset('iris')
```

```python
#scatter plot
sns.scatterplot(x=tips['total_bill'],y=tips['tip'],hue=tips['sex'],size=df['size'])
```

```python
#bar plot
sns.barplot(x=titanic['Pclass'],y=titanic['Age'],hue=titanic['Sex'])
```

```python
#box plot
sns.boxplot(x='Age',data=titanic)
```

```python
#heatmap cate-cate

sns.heatmap(pd.crosstab(titanic['Pclass'],titanic['Survived']))
```

```python
titanic.groupby('Pclass',).mean(numeric_only=True)['Survived'].plot(kind='bar')
```

```python
titanic.groupby('Sex').mean(numeric_only=True)['Survived'].plot(kind='bar')
```

```python
sns.clustermap(pd.crosstab(titanic['Parch'],titanic['Survived']))
```

```python
#pairplot
sns.pairplot(iris,hue='species')
```

```python
#lineplot num-num

flights.head()
new =flights.groupby('year').sum(numeric_only=True).reset_index()
```

```python
sns.lineplot(x=new['year'],y=new['passengers'])
```

```python
sns.heatmap(flights.pivot_table(values='passengers',index='month',columns='year'))
```

```python
sns.clustermap(flights.pivot_table(values='passengers',index='month',columns='year'))
```
